```python
import zipfile
import os

zip_file_path = '/content/API_SP.POP.TOTL_DS2_en_csv_v2_360546.zip'
extract_dir = '/content'

with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

print(f"Files extracted to: {extract_dir}")
print("List of files in the extract directory:")
print(os.listdir(extract_dir))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (uploaded in your Colab session)
df = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_360546.csv', skiprows=4)

# View available columns
print(df.columns)

# Example: Plotting the distribution of population for top 10 countries for the year 2023
df['2023'] = pd.to_numeric(df['2023'], errors='coerce')
top_10 = df.sort_values('2023', ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_10['Country Name'], y=top_10['2023'])
plt.title('Top 10 Most Populous Countries in 2023')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.show()
```
